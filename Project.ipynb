```python
import numpy as np

# Function to calculate maximum profit from stock prices
def max_profit(avg_prices):
    max_profit = 0
    for i in range(len(avg_prices)):
        for j in range(i + 1, len(avg_prices)):
            if avg_prices[j] > avg_prices[i]:
                profit = avg_prices[j] - avg_prices[i]
                max_profit = max(max_profit, profit)
    return max_profit

# Function to assess reliability based on historical data
def assess_reliability(avg_prices):
    changes = []
    for i in range(len(avg_prices) - 1):
        if avg_prices[i] != 0:
            change = abs(avg_prices[i + 1] - avg_prices[i]) / avg_prices[i]
            changes.append(change)
    average_change = np.mean(changes)
    if average_change <= 0.2:
        return "Reliable"
    else:
        return "Not Reliable"

# User Input Section
companies = []
avg_prices_list = []

for i in range(3):
    company_name = input(f"Enter name of Company {i + 1}: ")
    companies.append(company_name)

    avg_prices = []

    print(f"Enter historical average stock prices per day for {company_name}:")
    for day in range(7):
        while True:
            try:
                avg_price = float(input(f"Enter average stock price for Day {day + 1}: "))
                avg_prices.append(avg_price)
                break
            except ValueError:
                print("Please enter a valid number.")

    avg_prices_list.append(avg_prices)

while True:
    try:
        risk_tolerance = float(input("Enter your risk tolerance percentage (0-100): "))
        if 0 <= risk_tolerance <= 100:
            break
        else:
            print("Please enter a value between 0 and 100.")
    except ValueError:
        print("Please enter a valid number.")

# Data Analysis
max_profits = []
reliability = []

for i in range(3):
    max_profit_company = max_profit(avg_prices_list[i])
    max_profits.append(max_profit_company)

    reliability_status = assess_reliability(avg_prices_list[i])
    reliability.append(reliability_status)

# Recommendation
if risk_tolerance < 25:
    min_losses_index = np.argmin([sum(prices) / len(prices) for prices in avg_prices_list])
    recommended_company = companies[min_losses_index]
    print("\nRecommendation:")
    print(f"The most reliable company with the least average losses is: {recommended_company}")

elif 25 <= risk_tolerance < 50:
    min_losses_index = np.argmin([sum(prices) / len(prices) for prices in avg_prices_list])
    min_losses_company = companies[min_losses_index]
    max_profit_min_losses = max_profits[min_losses_index]

    # Initialize variables to track better options
    better_company = None
    better_profit = 0

    # Finding a company that can pay better than the one with minimum losses but should be reliable
    for i in range(3):
        if reliability[i] == "Reliable" and max_profits[i] > max_profit_min_losses:
            if max_profits[i] > better_profit:
                better_company = companies[i]
                better_profit = max_profits[i]

    if better_company:
        print("\nRecommendation:")
        print(f"The company that can pay better than {min_losses_company} and is reliable is: {better_company}")
    else:
        print("\nRecommendation:")
        print("No suitable company found within the reliability criteria.")

else:  # risk_tolerance >= 50
    max_profit_index = np.argmax(max_profits)
    recommended_company = companies[max_profit_index]
    print("\nRecommendation:")
    print(f"The company with the largest max profit is: {recommended_company} (even if non-reliable)")

print("\nCompany Reliability Assessment:")
for i in range(3):
    print(f"{companies[i]}: {reliability[i]}")
```
